```python
import pandas as pd
import numpy as np
import os
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
train_df = pd.read_csv('MB train.csv')
train_df.head()
```

```python
test_df = pd.read_csv('MB test.csv')
test_df.head()
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
#combine test and trin datasets in one datasets
train_df['source'] = 'train_df'
test_df['source'] = 'test_df'
data = pd.concat([train_df , test_df], ignore_index = True)
print(data.shape)
```

```python
pd.options.display.float_format = '{:.4f}'.format
variance = data.var()
variance = variance.reset_index()
variance.columns = ["ID","values"]
variance= variance.sort_values("values",ascending=1)
variance
```

```python
# Now,we will remove the variance 0 and low variance < 0.2
var = variance.loc[variance['values']<0.2 , 'ID']
data1 = data.drop(var,axis = 1)
data1.drop('ID',axis = 1,inplace = True)
data1.head()
```

```python
data1.shape
```

#### Check for null and unique values for test and train sets.

```python
#Missing Values in data
data1[data1.isnull().any(axis=1)]
```

```python
data1.isnull().sum()
```

```python
data1.describe()
```

```python
#fill the missing values by mean
data1['y'].fillna(data1['y'].mean(), inplace=True)
```

```python
data1.isnull().sum()
```

###Now, there is no missing values in any columns

```python
# Check for unique values in the data set
print(data1.nunique())
```

```python
data1.dtypes
```

```python
#Code all categorical variables as numeric using 'label Encoder' from sklearn's ptreprocessing module
#Initialize the Label Encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = data1.select_dtypes(include=['object']).columns.tolist()
for col in cat_cols:
    print(f"Unique values in {col} before encoding: {data1[col].unique()}")
    data1[col] = le.fit_transform(data1[col].astype(str))
    print(f"Unique values in {col} after encoding: {data1[col].unique()}")
```

```python
# Check for unique values in the data set
print(data1.nunique())
```

##### There are several techniques for dimensionality reduction, but one common technique is Principal Component Analysis (PCA). PCA is a technique that transforms a set of correlated variables into a smaller set of uncorrelated variables called principal components.

```python
## Perform PCA on the numerical columns
from sklearn.decomposition import PCA
num_cols = data1.select_dtypes(include = [np.number]).columns.tolist()
pca = PCA(n_components = 5)
data_pca = pd.DataFrame(pca.fit_transform(data1[num_cols]))
data_pca.columns = ['PC1','PC2','PC3','PC4','PC5']
data_pca.index = data1.index
```

```python
# Combine the PCA components with the categorical columns
data2 = pd.concat([data_pca, data1[cat_cols]], axis=1)
```

```python
# Check for unique values in the data set
print(data2.nunique())
```

```python
# Print explained variance ratio of the first two principal components
print(pca.explained_variance_ratio_[:5])
```

```python
# Get the explained variance ratio of all components
variance_ratio = pca.explained_variance_ratio_
variance_ratio
```

```python
# Get the cumulative explained variance ratio of all components
cumulative_variance_ratio = np.cumsum(variance_ratio)
cumulative_variance_ratio
```

```python
data1.head()
```

We can see that the cumulative explained variance ratio reaches a certain threshold i.e. 89.65% ,the principal components are capturing a sufficient amount of the variance in the data and that further dimensionality reduction may not be necessary.

##### Predict your data using XGBoost

```python
#segregating ind and dep var
X = data1.drop("y",axis=1)
y = data1.loc[:,"y"]
# Splitting
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y,test_size = 0.3, random_state = 123)
# Instantiation
import xgboost as xg
xgb_r = xg.XGBRegressor(objective ='reg:squarederror',n_estimators = 10, seed = 123)
# Fitting the model
xgb_r.fit(X_train,y_train)
# Predict the model
y_pred = xgb_r.predict(X_test)
###### mape
d = pd.DataFrame()
d["y_test"] = y_test
d["y_pred"] = y_pred
d["mp"] = abs((d["y_test"]- d["y_pred"])/d["y_test"])
(d.mp.mean())*100#mape
```

```python
# Evaluate the performance of the model on the test data
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

hence, the accuracy of this model is about 94% which is very good.
